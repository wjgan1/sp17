```python
train = train.drop('Unnamed: 0',axis=1)
```

```python
wine_labels = train['12']
```

```python
wine_labels
```

```python
wine_features = train.drop('12',axis=1)
```

```python
wine_features
```

```python
len(wine_features)
```

```python
wine_train_features = wine_features[0:4800]
```

```python
wine_train_features
```

```python
wine_train_labels = wine_labels[0:4800]
```

```python
wine_validation_features = wine_features[4800:]
```

```python
wine_validation_features
```

```python
wine_validation_labels = wine_labels[4800:]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
sizes = [100,200,500,1000,2000,4800]
```

```python
accuracy1 = []
```

```python
myLR1 = LogisticRegression()
```

```python
for n in sizes:
    myLR1.fit(X=wine_train_features[0:n],y=wine_train_labels[0:n])
    score = myLR1.score(X=wine_validation_features,y=wine_validation_labels)
    accuracy1.append(score)
```

```python
accuracy1
```

```python
plt.plot(sizes,accuracy1)
```

```python
plt.show()
```

```python
#Default penalty for Logistic Regression turns out to be L2,
#so we already plotted that
myLR2 = LogisticRegression(penalty='l1')
```

```python
accuracy2 = []
```

```python
for n in sizes:
    myLR2.fit(X=wine_train_features[0:n],y=wine_train_labels[0:n])
    score = myLR2.score(X=wine_validation_features,y=wine_validation_labels)
    accuracy2.append(score)
```

```python
accuracy2
```

```python
plt.plot(sizes,accuracy1)
```

```python
plt.plot(sizes,accuracy2)
```

```python
#The first one we plotted was L2, despite variable names
plt.legend(['L2','L1']) 
```

```python
plt.show()
```

```python
#L1 did better
```
