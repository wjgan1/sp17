```python
#Moving on to the kNN classifier
def kNN(k,X,Y,dist):
    def predict(x):
        #We're just going to brute force
        dists = [(dist(x,x_i),y) for (x_i,y) in zip(X,Y)]
        dists.sort(key=(lambda x: x[0]))
        kneighbors = dists[:k]
        votes = {}
        for k_i, y_i in kneighbors:
            if y_i not in votes:
                votes[y_i] = 0
            else:
                votes[y_i] += 1
        max_label = max(votes,key=votes.get)
        return max_label
    return predict
```

```python
def dist(x1,x2):
    return sum(np.square(x1-x2))
```

```python
#Simply evaluates how many were classified correctly
def evaluate(predict,X,Y):
    correct, total = 0, len(X)
    for x,y in zip(X,Y):
        if predict(x) == y:
            correct += 1
    return correct/total
```

```python
#Attempted to be general but really only works for kNN
#Eventually should replace with functions from sklearn,
#which can also be used to display more stats
def kfold(k,X,Y,model=None,kN=5):
    avg_score = 0
    bsz = len(X)//k
    for i in range(0,k):
        train_x, train_y = np.concatenate((X[0:bsz*i],X[bsz*(i+1):])), np.concatenate((Y[0:bsz*i],Y[bsz*(i+1):]))
        validation_x, validation_y = X[bsz*i:bsz*(i+1)], Y[bsz*i:bsz*(i+1)]
        predict = model(kN,train_x,train_y,dist)
        avg_score += evaluate(predict,validation_x,validation_y)
        #print(avg_score)
    avg_score /= k
    return avg_score
```

```python
kfold(5,X,Y,kNN,5)
```
